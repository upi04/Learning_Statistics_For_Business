## Case 1 - Sales Transaction

**Context**<br>
E-commerce has become a new channel to support businesses development. Through e-commerce, businesses can get access and establish a wider market presence by providing cheaper and more efficient distribution channels for their products or services. E-commerce has also changed the way people shop and consume products and services. Many people are turning to their computers or smart devices to order goods, which can easily be delivered to their homes.

**Content**<br>
This is a sales transaction data set of UK-based e-commerce (online retail) for one year in 2019. This London-based shop has been selling gifts and homewares for adults and children through the website since 2007. Their customers come from all over the world and usually make direct purchases for themselves. There are also small businesses that buy in bulk and sell to other customers through retail outlet channels.

**Description of Variables**

- TransactionNo (categorical): a six-digit unique number that defines each transaction. The letter “C” in the code indicates a cancellation.
- Date (numeric): the date when each transaction was generated.
- ProductNo (categorical): a five or six-digit unique character used to identify a specific product.
- Product (categorical): product/item name.
- Price (numeric): the price of each product per unit in pound sterling (£).
- Quantity (numeric): the quantity of each product per transaction. Negative values related to cancelled transactions.
- Country (categorical): name of the country where the customer resides.

4. How was the sales trend over the months?

```python
# change data type of 'Date' to datetime to processed datetime variable easily
dateFormat = '%Y-%m-%d'
df['Date'] = df['Date'].astype(str)
non_date_rows = df[~df['Date'].str.match(r'\d{4}-\d{2}-\d{2}', na=False)]
print(non_date_rows)

# Create New Columns : Month
df['Month'] = pd.DatetimeIndex(df['Date']).month
```

```python
sns.countplot(x = df["Month"], color="green").set(title='Transaction Each Month in 2019')
```

*   Most transactions occurred in November
*   The fewest transactions occurred in December
*   Most transactions occur in autumn season

To find the mode of the data, we can use function `mode` from `statistics` library



```python
mode(df['Month'])
```

```python
import datetime as dt

# Get unique months
month = sorted(df['Month'].unique())
df['Month'] = df['Month'].fillna(1).astype(int)  # Fill NaN with 1 (January) and then convert to integers


# Convert month number to month name
listMonth = [dt.datetime(2021, int(i), 1).strftime("%b") for i in month]

# Sum TotalMoney by month
listTotalMoney = [df['TotalMoney'].loc[df['Month'] == i].sum() for i in month]

# Create the DataFrame
dfMonth = pd.DataFrame({'MonthName': listMonth, 'TotalMoney': listTotalMoney})

# Display the result
print(dfMonth.head())
```

```python
plt.figure(figsize = (10,6))
plt.plot(dfMonth['MonthName'], dfMonth['TotalMoney'], marker = 's',alpha = 0.8)
plt.xlabel('Month')
plt.ylabel('Total Money')
plt.title('Total Money Each Months in 2019')
plt.show()
```

```python
# Make DataFrame of Month and Quantity

month = sorted(df['Month'].unique())
listMonth = [dt.datetime.strptime(str(i), "%m").strftime("%b") for i in month]
listQuantity = [df['Quantity'].loc[df['Month']==i].sum() for i in month]
dfMonthQuantity = pd.DataFrame({'MonthName' : listMonth,'Quantity' : listQuantity})

plt.figure(figsize = (10,6))
plt.plot(dfMonthQuantity['MonthName'], dfMonthQuantity['Quantity'], marker = 's',alpha = 0.8)
plt.xlabel('Month')
plt.ylabel('Quantity')
plt.title('Quantity Each Months in 2019')
plt.show()
```

In comparison to other seasons, we can observe from the figure "*Total money Each Month in 2019*" that **sales are highest in the autumn**. this is reinforced by the large quantity transacted each month (from "*Quantity Each Month in 2019*").

Advertising and promotion should be done to optimize revenue while adding demand during this season. Further investigation should be carried out to find out why the product is not selling well in other seasons and make an informed decision.

## Case 2 - Fast Food Marketing Campaign
**Scenario**<br>
A fast-food chain plans to add a new item to its menu. However, they are still undecided between three possible marketing campaigns for promoting the new product. In order to determine which promotion has the greatest effect on sales, the new item is introduced at locations in several randomly selected markets. A different promotion is used at each location, and the weekly sales of the new item are recorded for the first four weeks.

**Goal**<br>
Describe the characteristics of the Data

**Description of Variables**<br>
1. MarketID: unique identifier for market
2. MarketSize: size of market area by sales
3. LocationID: unique identifier for store location
4. AgeOfStore: age of store in years
5. Promotion: one of three promotions that were tested
6. week: one of four weeks when the promotions were run
7. SalesInThousands: sales amount for a specific LocationID, Promotion, and week
